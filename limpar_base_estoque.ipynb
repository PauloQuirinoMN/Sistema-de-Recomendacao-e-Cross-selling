```python
df_estoque = base_estoque[["Código", "Produto", "Código da Marca", "Marca", "Código da categoria", "Categoria", "Preço de custo", "Quantidade estoque"]].copy()
```

```python
df_estoque.isnull().sum()
```

```python
colunas_num = ['Código', 'Código da Marca', 'Código da categoria', 'Quantidade estoque', 'Preço de custo']
for coluna in colunas_num:
    qtd = (df_estoque[coluna] == 0).sum()
    print(f"{coluna}: {qtd} valores zerados")
```

```python
df_estoque['Quantidade estoque'] = df_estoque['Quantidade estoque'].fillna(0)
```

```python
zeros = ((df_estoque["Preço de custo"] == 0).sum()*100 / df_estoque["Código"].count())
print(f"{zeros:.2f} % de valores zerados na coluna de preço de custo")
```

```python
cod_prod_uni = df_estoque["Código"].nunique()
pro_unicos = df_estoque["Produto"].nunique()
print(f"{cod_prod_uni} códigos únicos\n{pro_unicos} produtos únicos\n{cod_prod_uni-pro_unicos} diferença")
```

```python
cod_cat_uni = df_estoque["Código da categoria"].nunique()
cat_unicos = df_estoque["Categoria"].nunique()
print(f"{cod_cat_uni} códigos de categorias únicos\n{cat_unicos} categorias únicas\n{cod_cat_uni-cat_unicos} deferença")
```

```python
cod_marc_inic = df_estoque["Código da Marca"].nunique()
marc_unicas = df_estoque["Marca"].nunique()
print(f"{cod_marc_inic} códigos únicos das marcas\n{marc_unicas} Marcas únicas\n{cod_marc_inic-marc_unicas} diferença")
```

```python
# Caso 1: Um produto com múltiplos códigos
produtos_duplicados = df_estoque.groupby("Produto")["Código"].nunique().sort_values(ascending=False)
print(produtos_duplicados[produtos_duplicados > 1].head(20))
```

```python
# Primeiro, identifique os produtos com maior variação de códigos:
# Agrupar produtos com +1 código
produtos_multiplos_codigos = df_estoque.groupby('Produto')['Código'].nunique().reset_index(name='Qtd_Códigos')
top_problemas = produtos_multiplos_codigos[produtos_multiplos_codigos['Qtd_Códigos'] > 1].sort_values('Qtd_Códigos', ascending=False)

# Mesclar com categoria/marca para análise
top_problemas = top_problemas.merge(
    df_estoque[['Produto', 'Categoria', 'Marca']].drop_duplicates(),
    on='Produto',
    how='left'
)
display(top_problemas.head(10))
```

```python
top_problemas.to_excel("Produtos_codigos_duplicados.xlsx", index=False)
```

```python
# % de produtos com múltiplos códigos por categoria
analise_categoria = top_problemas.groupby('Categoria')['Produto'].count() / df_estoque['Categoria'].nunique() * 100
analise_categoria.sort_values(ascending=False).head(10)
```

```python
# Top marcas problemáticas
analise_marca = top_problemas['Marca'].value_counts(normalize=True).head(10)
print(analise_marca)
```
